```python
# packages
import numpy as np
import pandas as pd
import altair as alt

# raw data
air_raw = pd.read_csv('air-quality.csv')
cbsa_info = pd.read_csv('cbsa-info.csv')

## PART I
##########
data = pd.merge(air_raw, cbsa_info, how = 'left', on = 'CBSA')

def get_element(l, index): # helper function to split up city and state, to figure out unique states/territories
    return l[index]

data['cities_covered'] = data['Core Based Statistical Area'].str.split(',').apply(get_element, index = 0)
data['state_terr_covered'] = data['Core Based Statistical Area'].str.split(',').apply(get_element, index = 1)
data['state_terr_covered'].unique().shape # figure out how many unique states/territories CBSAs cover

data.isna().sum(axis = 0) # figure out which variable has the most missing values --> no missing values

## PART II (continued in subsequent cells)
##########

# Has PM 2.5 air pollution improved in the U.S. on the whole since 2000?
# Scatter plots of PM2.5 over the years, using weighted average mean
yrs = range(2000, 2020)
yrs = list(map(str, range(2000,2020)))

plot_means = data[(data['Pollutant'] == 'PM2.5') & (data['Trend Statistic'] == 'Weighted Annual Mean')].drop(
             columns = ['Pollutant', 'Number of Trends Sites', 'Core Based Statistical Area', 'state_terr_covered']
             ).melt(
             id_vars = ['CBSA'], value_vars = yrs, var_name = 'Year', value_name = 'Concentration'
             ).reset_index()

alt.Chart(plot_means).mark_point().encode( # your turn here
    x = alt.X('CBSA', scale = alt.Scale(type = 'pow', zero = False)), # your turn here
    y = alt.Y('Concentration', scale = alt.Scale(type = 'log')
)).facet(column = 'Year')
```

```python
# city with the best improvement (2000-2019)

cities_diff = data[(data['Pollutant'] == 'PM2.5') & (data['Trend Statistic'] == 'Weighted Annual Mean')].drop(
              columns = ['Pollutant', 'Number of Trends Sites']
              ).melt(
              id_vars = ['Core Based Statistical Area', 'CBSA', 'cities_covered', 'state_terr_covered'], value_vars = yrs, var_name = 'Year', value_name = 'Concentration'
              ).reset_index()

cities_diff = cities_diff[(cities_diff.Year == '2000') | (cities_diff.Year == '2019')].pivot(index = 'Core Based Statistical Area', columns = 'Year', values = 'Concentration')
cities_diff['improvement00-19'] = cities_diff['2019'] - cities_diff['2000']
cities_diff[cities_diff['improvement00-19'] == cities_diff['improvement00-19'].min()]
```

```python
# territory with the best improvement (2000-2019)

state_terr_diff = data[(data['Pollutant'] == 'PM2.5') & (data['Trend Statistic'] == 'Weighted Annual Mean')].drop(
              columns = ['Pollutant', 'Number of Trends Sites']
              ).melt(
              id_vars = ['state_terr_covered', 'CBSA'], value_vars = yrs, var_name = 'Year', value_name = 'Concentration'
              ).reset_index()

state_terr_diff = state_terr_diff[(state_terr_diff.Year == '2000') | (state_terr_diff.Year == '2019')].groupby(
    ['state_terr_covered', 'Year']
    ).mean().drop(columns = ['index', 'CBSA']).reset_index(
    ).pivot(index = 'state_terr_covered', columns = 'Year', values = 'Concentration')

state_terr_diff['improvement00-19'] = state_terr_diff['2019'] - state_terr_diff['2000']
state_terr_diff[state_terr_diff['improvement00-19'] == state_terr_diff['improvement00-19'].min()]
```

```python
# find state with the best improvement (2000-2019)

state_terr_diff.sort_values('improvement00-19', ascending = True)
```

```python
# find PM 2.5 concentration in 2019 for SF-Oakland-Hayward

data = data[(data.cities_covered == 'San Francisco-Oakland-Hayward') & (data.Pollutant == 'PM2.5') & (data['Trend Statistic'] == 'Weighted Annual Mean')].melt(
      id_vars = ['cities_covered'], value_vars = yrs, var_name = 'Year', value_name = 'Concentration'
      )

data[data.Year == '2019']
```

## Notes on merging (keep at bottom of notebook)

To combine datasets based on shared information, you can use the `pd.merge(A, B, how = ..., on = SHARED_COLS)` function, which will match the rows of `A` and `B` based on the shared columns `SHARED_COLS`. If `how = 'left'`, then only rows in `A` will be retained in the output (so `B` will be merged *to* `A`); conversely, if `how = 'right'`, then only rows in `B` will be retained in the output (so `A` will be merged *to* `B`).

A simple example of the use of `pd.merge` is illustrated below:

```python
# toy data frames
A = pd.DataFrame(
    {'shared_col': ['a', 'b', 'c'], 
    'x1': [1, 2, 3], 
    'x2': [4, 5, 6]}
)

B = pd.DataFrame(
    {'shared_col': ['a', 'b'], 
    'y1': [7, 8]}
)
```

```python
A
```

```python
B
```

Below, if `A` and `B` are merged retaining the rows in `A`, notice that a missing value is input because `B` has no row where the shared column (on which the merging is done) has value `c`. In other words, the third row of `A` has no match in `B`.

```python
# left join
pd.merge(A, B, how = 'left', on = 'shared_col')
```

If the direction of merging is reversed, and the row structure of `B` is dominant, then the third row of `A` is dropped altogether because it has no match in `B`.

```python
# right join
pd.merge(A, B, how = 'right', on = 'shared_col')
```
